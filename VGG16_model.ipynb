## Y data 생성

```python
## Y data 생성
t=list()

for _ in range(int(data.shape[0]/300)):
    for j in range(300):
        t.append(_)
t=np.array(t)

targets = np.array(pd.get_dummies(t)).astype('int')
```

## train test split

```python
# train test split
X_train, X_test, y_train, y_test = train_test_split(data,targets)
```

# 모델 생성
- Pretrained CNN : VGG16 사용

```python
from keras.applications import VGG16
# vgg16은 224x224 사이즈 이미지를 입력으로 받아 학습된 상태인데
# 그보다 작은 사이즈도 된다. 혹시나 잘 안되면 224x224에 맞춰보자.
pre_trained_model = VGG16(weights='imagenet',  # 이미지넷의 데이터로 1000개 클래스 분류에 대해 학습된 가중치
                          include_top=None,   # VGG16에서 5번째 맥스풀링 이후의 fc layer는 무시한다.
                          input_shape=(150,150, 3))
model = models.Sequential()
model.add(pre_trained_model)
model.add(layers.Flatten())
model.add(layers.Dropout(0.3))
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dropout(0.3))
model.add(layers.Dense(30, activation='softmax'))

# optimizers.Adam
model.compile(optimizer=optimizers.Adam(lr=0.005, beta_1=0.9, beta_2=0.999, 
                                         decay=1e-6, amsgrad=False),
              loss='categorical_crossentropy',
              metrics=['acc'])
model.summary()
```

## min-max scaling

```python

X_train = X_train/255.0
X_test = X_test/255.0
# y_train = y_train.astype('float32')
# y_test = y_test.astype('float32')
```

```python
model.fit(x=X_train,y=y_train,batch_size=10,epochs=10)
```

```python
test_acc,test_loss = model.evaluate(X_test,y_test,workers=4)
```
